```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.neural_network import MLPClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import cross_val_score
from sklearn.decomposition import PCA
import joblib
```

```python
df = pd.read_csv("data/loan_prediction_dataset.csv")
df
```

```python
le = LabelEncoder()
df["Employment_Status"] = le.fit_transform(df["Employment_Status"])
df
```

```python
X = df.drop(columns = ["Loan_Approved"])
y = df["Loan_Approved"]
```

```python
pca = PCA(n_components = 5)
X = pca.fit_transform(X)
```

```python
pca.explained_variance_ratio_
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0, stratify = y)
```

```python
mlp = MLPClassifier()
scores = cross_val_score(mlp, X_train, y_train, cv = 10, scoring = "accuracy")
scores.mean()
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(200,100),activation='relu', learning_rate='adaptive', max_iter=1000, alpha=0.0001, random_state=42)
scores = cross_val_score(mlp, X_train, y_train, cv = 10, scoring = "accuracy")
scores.mean()
```
